## Task 2 - NLP

The dataset CENSUS-NER-Name-Email-Address-Phone is available at [Hugging Face](https://huggingface.co/datasets/Josephgflowers/CENSUS-NER-Name-Email-Address-Phone). Write a single regular expression that will find email addresses in various formats in the 'user' column, for example:
- user@domain.com
- user@domain.co.uk
- user123@domain.com
- user+tag@domain.com

Count the occurrences of domain names. Present the results in a bar chart. Solutions other than exclusively one regular expression will not be considered.

```python
# import necesary libraries
from huggingface_hub import hf_hub_download
import pandas as pd
from matplotlib import pyplot as plt
import re
```

```python
# Load dataset (using hugging face docs)
REPO_ID = "Josephgflowers/CENSUS-NER-Name-Email-Address-Phone"
FILENAME = "FMCSA_CENSUS1_2016Sep_formatted_output.csv"

dataset = pd.read_csv(
    hf_hub_download(repo_id=REPO_ID, filename=FILENAME, repo_type="dataset")
)
```

```python
#show format of dataset
dataset[0:3]
```

```python
#test valuses for regex
pattern = "[a-zA-Z0-9_+]+@[a-zA-Z0-9.]+"
regex_test_true = ["user@domain.com", "user@domain.co.uk", "user123@domain.com", "user+tag@domain.com"]
regex_test_false = ["@me", "domain.com", "user@", "", " "]
for i in regex_test_true:
    print(re.search(pattern,i))
print("------------")
for i in regex_test_false:
    print(re.search(pattern,i), end=' ')
# this regex is lightweight and good enough to use for 1M simple data
```

```python
# search and count email domains
occurances = {}
list_of_top_domains = []
for i in range(len(dataset['user'])):
    mail = re.search(pattern,dataset['user'][i])
    if mail is not None:
        domain = mail[0].split("@")[1]
        if domain in occurances.keys():
            occurances[domain]+=1
        else:
            occurances[domain]=1
```

Sometimes the nomenclature “emailserver.domain” is used, so to standardize what is used in the project as above (source in markdown)

```python
# For bar choose top domains and rest show as 'other'
names = ['OTHER']
values = [0]
occurances = dict(sorted(occurances.items(), key=lambda x:x[1],reverse=True))
for i in occurances.keys():
    if occurances[i]>500:
        print(i, occurances[i], end='    ')
        names.append(i)
        values.append(occurances[i])
    else:
        values[0] +=occurances[i]
```

```python
fig = plt.figure(figsize = (10, 5))
plt.title('Domain name occurrences')
plt.ylabel('Log scale number of domain')
plt.xticks(rotation=80)
plt.bar(names, values, width = 0.4,log=True, color = (0.5,0.1,0.5,0.6))
```

To avoid too small bars in the graph, a logarithmic scale was used
